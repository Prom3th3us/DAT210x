```python
#ordinal categorical data
income_ordered = ['a','b','c','d','e']
df = pd.DataFrame({'income':['a','e','c','a','e', 'niegevonde']})
```

```python
df
```

```python
df.income = df.income.astype('category',ordered = True, categories = income_ordered).cat.codes
```

```python
df
```

```python
#nominal categorical data: mapping entries to ints (alhphabetically)
df = pd.DataFrame({'income':['a','e','c','a','e', 'niegevonde']})
```

```python
df.income = df.income.astype('category').cat.codes
```

```python
df
```

```python
# other approach for nominal categorical features: dummy variables
df = pd.DataFrame({'income':['a','e','c','a','e', 'niegevonde']})
```

```python
df = pd.get_dummies(df,columns = ['income'])
```

THe SciKitLearn Coutn vectorizer is a very usefull library for feature representation of textual data. It allows to count words or characters, can be configured to ignore most common words and can also be fed your own tokenisation functions

```python
# Feature representation of textual data using the SciKit Learn CountVectorizer
from sklearn.feature_extraction.text import CountVectorizer
corpus = ['Authman ran faster than Harry because he is an athlete.',
         'Authman and Harry ran faster and faster.']
bow = CountVectorizer()
X = bow.fit_transform(corpus) #Sparse Matrix, compact memory representation
print(bow.get_feature_names())
print(X.toarray())
```
